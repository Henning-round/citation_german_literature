```python
folder_path = 'grobid_results/'
import glob
list_authors = []

for filename in tqdm(glob.glob(os.path.join(folder_path, '*.xml'))):
    with open(filename, 'r') as tei:
        soup = BeautifulSoup(tei, 'lxml')
        persnames = soup.find_all("persname")
        list_authors += list(set([" ".join(names.strings).lower() for names in persnames]))
        
```

```python
collection = Counter(list_authors)
```

```python
len(list_authors)
```

```python
order = {k: v for k, v in sorted(collection.items(), key=lambda item: item[1], reverse=True)}
```
